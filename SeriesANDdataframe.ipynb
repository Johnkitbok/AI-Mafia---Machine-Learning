# MatplotLip

```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,5]
y=[2,3,6]
#for plotting this x,y point
```

```python
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
```

```python
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.plot(x,y)
plt.title('codingclub')
plt.show()
```

```python
#line 1
x1 = [1,2,3]
y1 = [2,4,1]

#line 2
x2 = [1,2,3]
y2 = [4,1,3]

#plotting the line 1 points
plt.plot(x1,y1, label = "line 1")

#plotting the line 2 points
plt.plot(x2, y2, label = "line 2")
plt.legend()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
```
